```python
import pandas as pd

df = pd.read_csv("df_cleaned.csv')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df.head()
```

```python
month_mapping = {
    "jan": 1,
    "feb": 2,
    "mar": 3,
    "apr": 4,
    "may": 5,
    "jun": 6,
    "jul": 7,
    "aug": 8,
    "sep": 9,
    "oct": 10,
    "nov": 11,
    "dec": 12,
}
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder

oc = OneHotEncoder(drop="first", sparse_output=False)
ed_cat = [["primary", "secondary", "tertiary"]]
month_cat = [list(month_mapping.keys())]
oe = OrdinalEncoder(categories=ed_cat)
oe_month = OrdinalEncoder(categories=month_cat)
```

```python
trf = ColumnTransformer(
    transformers=[
        ("OHE", oc, ["job", "marital", "contact", "poutcome"]),
        ("Ordinal_Encoding", oe, ["education"]),
        ("Ordinal_month", oe_month, ["month"]),
    ],
    remainder="passthrough",
)
```

```python
data = trf.fit_transform(df)
```

```python
col = trf.get_feature_names_out()
```

```python
ml_df = pd.DataFrame(data=data, columns=col)
```

```python
lc = LabelEncoder()
ml_df["remainder__y"] = lc.fit_transform(ml_df["remainder__y"])
```

```python
ml_df.drop(columns="remainder__Unnamed: 0", inplace=True)
```

```python
ml_df.to_csv("machine_learning.csv")
```
